```python
fechas = list(df_full['DateCreated'].unique())
df = pd.DataFrame(fechas, index=range(len(fechas)),columns=['Fecha'])
df['Fecha'] = df['Fecha'].apply(lambda x: x.strftime("%Y-%m-%d"))
df['Fecha_def'] = df_full['DateCreated'].unique()
df['Año-Mes'] = df['Fecha'].str.slice(0, 7)
df['Año-Mes'] = df['Año-Mes'].apply(lambda x: x.replace("-",""))
df
```

```python
df['Weekday'] = df['Fecha_def'].apply(lambda x: x.isoweekday())
df['Week'] = df['Fecha_def'].apply(lambda x: x.isocalendar()[1])
df
```
